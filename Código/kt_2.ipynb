```python
# esto es con 4 restricciones de desigualdad

import numpy as np
from scipy.optimize import minimize

# Objective function to minimize (since we are minimizing)
def objective(xy):
    x, y = xy
    return -((x - 11)**2 + (y - 11)**2 ) 

# Constraints definitions
def constraint_x(xy):
    x, y = xy
    return x - 4  # x >= 4

def constraint_y(xy):
    x, y = xy
    return y - 4  # y >= 4

def constraint_x1(xy):
    x, y = xy
    return 10 - x  # x <= 10

def constraint_y1(xy):
    x, y = xy
    return 10 - y  # y <= 10

# Collect constraints
constraints = [
    {'type': 'ineq', 'fun': constraint_x},   # x >= 1
    {'type': 'ineq', 'fun': constraint_y},   # y >= 1
    {'type': 'ineq', 'fun': constraint_x1},  # x <= 10
    {'type': 'ineq', 'fun': constraint_y1},  # y <= 10
]

# Define bounds for variables (x, y)
#bounds = [(1, 10), (1, 10)]  # x and y between 1 and 10

# Initial guess
initial_guess = [1.5, 1.5]

# Solve the optimization problem using 'trust-constr'
result = minimize(
    objective,
    initial_guess,
    method='trust-constr                                                                                            ',
    #bounds=bounds,
    constraints=constraints,
    options={'verbose': 1}
)

# Display results
if result.success:
    print("\nOptimal solution found:")
    print(f"x = {result.x[0]:.4f}, y = {result.x[1]:.4f}")
    print(f"Minimum value of f(x, y) = {result.fun:.4f}")

    # Extract Kuhn-Tucker multipliers (Lagrange multipliers)
    multipliers = result.v

    # Try accessing 'ineq' multipliers
    if hasattr(multipliers, 'ineq'):
        ineq_multipliers = multipliers.ineq
    elif isinstance(multipliers, dict) and 'ineq' in multipliers:
        ineq_multipliers = multipliers['ineq']
    else:
        # Assuming 'multipliers' is an array of multipliers for inequality constraints
        ineq_multipliers = multipliers

    print("\nInequality Multipliers (full):")
    print(ineq_multipliers)

    print("\nKuhn-Tucker Multipliers:")
    print("Inequality Multipliers:")
    for i, con in enumerate(constraints):
        multiplier_value = ineq_multipliers[i]

        # If multiplier_value is an array, extract scalar
        if isinstance(multiplier_value, np.ndarray):
            multiplier_value = multiplier_value.item()

        print(f"Constraint {i+1}: Multiplier = {multiplier_value:.4f}")
else:
    print("Optimization failed:", result.message)
```

```python
import numpy as np
from scipy.optimize import minimize

# Objective function to minimize (since we are minimizing)
def objective(xy):
    x, y = xy
    return -((x - 11)**2 + (y - 11)**2)

# Inequality constraints
def constraint_x(xy):
    x, y = xy
    return x - 3.9  # x >= 3.9

def constraint_y(xy):
    x, y = xy
    return y - 4  # y >= 4

def constraint_x1(xy):
    x, y = xy
    return 10 - x  # x <= 10

def constraint_y1(xy):
    x, y = xy
    return 10 - y  # y <= 10

# Equality constraint
def equality_constraint(xy):
    x, y = xy
    return x + y - 9  # x + y = 9

# Collect all constraints
constraints = [
    {'type': 'ineq', 'fun': constraint_x},
    {'type': 'ineq', 'fun': constraint_y},
    {'type': 'ineq', 'fun': constraint_x1},
    {'type': 'ineq', 'fun': constraint_y1},
    {'type': 'eq', 'fun': equality_constraint},  # Add equality constraint
]

# Initial guess
initial_guess = [1.5, 1.5]

# Solve the optimization problem using 'trust-constr'
result = minimize(
    objective,
    initial_guess,
    method='trust-constr',
    constraints=constraints,
    options={'verbose': 1}
)

# Display results
if result.success:
    print("\nOptimal solution found:")
    print(f"x = {result.x[0]:.4f}, y = {result.x[1]:.4f}")
    print(f"Objective function value at optimal solution = {result.fun:.4f}")

    # Try accessing Lagrange multipliers
    print("\nKuhn-Tucker Multipliers (Lagrange Multipliers):")
    if hasattr(result, 'v'):
        # Access inequality multipliers
        ineq_multipliers = result.v.ineq if hasattr(result.v, 'ineq') else result.v[0:len(constraints) - 1]
        print("Inequality Multipliers:", ineq_multipliers)

        # Access equality multipliers
        eq_multipliers = result.v.eq if hasattr(result.v, 'eq') else result.v[len(constraints) - 1:]
        print("Equality Multipliers:", eq_multipliers)
    else:
        print("Lagrange multipliers are not directly available in 'trust-constr'.")
else:
    print("Optimization failed:", result.message)
```

```python
# Print inequality constraints and their multipliers
print("\nInequality Multipliers:")
ineq_index = 0
for i, constraint in enumerate(constraints):
    if constraint['type'] == 'ineq':
        multiplier = ineq_multipliers[ineq_index][0] if isinstance(ineq_multipliers[ineq_index], np.ndarray) else ineq_multipliers[ineq_index]
        print(f"  Constraint {i + 1} ({constraint['fun'].__name__}): Multiplier = {multiplier:.6f}")
        ineq_index += 1

# Print equality constraints and their multipliers
print("\nEquality Multipliers:")
eq_index = 0
for i, constraint in enumerate(constraints):
    if constraint['type'] == 'eq':
        multiplier = eq_multipliers[eq_index][0] if isinstance(eq_multipliers[eq_index], np.ndarray) else eq_multipliers[eq_index]
        print(f"  Equality Constraint {i + 1} ({constraint['fun'].__name__}): Multiplier = {multiplier:.6f}")
        eq_index += 1
```
